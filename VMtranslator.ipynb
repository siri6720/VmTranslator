```python
class VMtranslator:
    
    import sys

    #set the self.input and self.output paths to generate the .asm file inside of the directory

    def __init__(self, VMfileLoc):

        path = VMfileLoc
        print("path:", path)

        # Open file into inputdata variable for reading it line by line
        self.inputdata = open(path,"r")
        # asmFile variable to append the codes line by line
        self.asmFile = ""
        # Replace .vm with .asm for output file name
        self.asmFileLoc = path.replace('.vm','.asm')
        # Create array of arryas with possible VM commads and assign the function names
        self.instructions = [['add', '0', self.add],
                    ['sub', '1', self.sub],
                    ['neg', '2', self.neg],
                    ['eq', '3', self.eq],
                    ['gt', '4', self.gt],
                    ['lt','5', self.lt],
                    ['and','6', self.f_and],
                    ['or','7', self.f_or],
                    ['not','8', self.f_not],
                    ['pop local','9', self.asm_pop_local],
                    ['pop argument','10', self.asm_pop_argument],
                    ['pop this','11', self.asm_pop_this],
                    ['pop that','12', self.asm_pop_that],
                    ['pop static','13', self.asm_pop_static],
                    ['pop pointer 0','14', self.asm_pop_pointer_0],
                    ['pop pointer 1','15', self.asm_pop_pointer_1],
                    ['pop temp','16', self.asm_pop_temp],
                    ['push local','17', self.asm_push_local],
                    ['push argument','18', self.asm_push_argument],
                    ['push this','19', self.asm_push_this],
                    ['push that','20', self.asm_push_that],
                    ['push constant','21', self.asm_push_constant],
                    ['push static','22', self.asm_push_static],
                    ['push pointer 0','23', self.asm_push_pointer_0],
                    ['push pointer 1','24', self.asm_push_pointer_1],
                    ['push temp','25', self.asm_push_temp]] 
        
    # in the following functions, we define the VM to asm translations so that
    # it can be written to a file
    
    #ASM PUSH COMMANDS
    def asm_push_constant(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[14:])
        self.asmFile = self.asmFile + ('D=A\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_local(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[11:])
        self.asmFile = self.asmFile + ('D=A\n@LCL\nA=D+M\nD=M\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_that(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[10:])
        self.asmFile = self.asmFile + ('D=A\n@THAT\nA=D+M\nD=M\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_argument(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[14:])
        self.asmFile = self.asmFile + ('D=A\n@ARG\nA=D+M\nD=M\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_this(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[10:])
        self.asmFile = self.asmFile + ('D=A\n@THIS\nA=D+M\nD=M\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_temp(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[10:])
        self.asmFile = self.asmFile + ('D=A\n@5\nA=D+A\nD=M\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_pointer_0(self, line, i):
        #self.asmFile = self.asFiasmFile + ('@'+line[10:])
        self.asmFile = self.asmFile + ('@3\nD=M\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_pointer_1(self, line, i):
        #self.asmFile = self.asFiasmFile + ('@'+line[10:])
        self.asmFile = self.asmFile + ('@4\nD=M\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def asm_push_static(self, line, i):
        self.asmFile = self.asmFile + ('@something.'+line[12:])
        self.asmFile = self.asmFile + ('D=M\n')
        self.asmFile = self.asmFile + ('@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    #ASM POP COMMANDS
    def  asm_pop_static(self, line, i):
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n')
        self.asmFile = self.asmFile + ('@something.'+line[11:])
        self.asmFile = self.asmFile + ('M=D\n')

    def  asm_pop_local(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[10:])
        self.asmFile = self.asmFile + ('D=A\n@LCL\nD=D+M\n@SP\nA=M\nM=D\nA=A-1\nD=M\nA=A+1\nA=M\nM=D\n@SP\nM=M-1\n')
        
    def  asm_pop_argument(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[13:])
        self.asmFile = self.asmFile + ('D=A\n@ARG\nD=D+M\n@SP\nA=M\nM=D\nA=A-1\nD=M\nA=A+1\nA=M\nM=D\n@SP\nM=M-1\n')

    def  asm_pop_this(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[9:])
        self.asmFile = self.asmFile + ('D=A\n@THIS\nD=D+M\n@SP\nA=M\nM=D\nA=A-1\nD=M\nA=A+1\nA=M\nM=D\n@SP\nM=M-1\n')
    
    def  asm_pop_that(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[9:])
        self.asmFile = self.asmFile + ('D=A\n@THAT\nD=D+M\n@SP\nA=M\nM=D\nA=A-1\nD=M\nA=A+1\nA=M\nM=D\n@SP\nM=M-1\n')

    def  asm_pop_temp(self, line, i):
        self.asmFile = self.asmFile + ('@'+line[9:])
        self.asmFile = self.asmFile + ('D=A\n@5\nD=D+A\n@SP\nA=M\nM=D\nA=A-1\nD=M\nA=A+1\nA=M\nM=D\n@SP\nM=M-1\n')
    
    def  asm_pop_pointer_0(self, line, i):
        self.asmFile = self.asmFile + ('@SP\nM=M-1\n@3\nM=A\nM=D\n')

    def  asm_pop_pointer_1(self, line, i):
        self.asmFile = self.asmFile + ('@SP\nM=M-1\n@4\nM=A\nM=D\n')

    #ASM ARITHMETICAL COMMANDS
    def add(self, line, i):
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n@SP\nM=M-1\nA=M\nM=D+M\n@SP\nM=M+1\n')

    def sub(self, line, i):
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n@SP\nM=M-1\nA=M\nM=M-D\n@SP\nM=M+1\n')

    def f_and(self, line, i):
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n@SP\nM=M-1\nA=M\nM=M&D\n@SP\nM=M+1\n')

    def f_or(self, line, i):
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n@SP\nM=M-1\nA=M\nM=M|D\n@SP\nM=M+1\n')

    def eq(self, line, i):
        n=str(i)
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n@SP\nM=M-1\nA=M\nD=D-M\n@JUMP'+n+'\nD,JEQ\nD=0\n@ENDING'+n+'\n0,JMP\n(JUMP'+n+')\nD=-1\n(ENDING'+n+')\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def lt(self, line, i):
        n=str(i)
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n@SP\nM=M-1\nA=M\nD=D-M\n@JUMP'+n+'\nD,JGT\nD=0\n@ENDING'+n+'\n0,JMP\n(JUMP'+n+')\nD=-1\n(ENDING'+n+')\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def gt(self, line, i):
        n=str(i)
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=M\n@SP\nM=M-1\nA=M\nD=D-M\n@JUMP'+n+'\nD,JLT\nD=0\n@ENDING'+n+'\n0,JMP\n(JUMP'+n+')\nD=-1\n(ENDING'+n+')\n@SP\nA=M\nM=D\n@SP\nM=M+1\n')

    def neg(self, line, i):
        n=str(i)
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=!M\nM=D+1\n@SP\nM=M+1\n')

    def f_not(self, line, i):
        n=str(i)
        self.asmFile = self.asmFile + ('@SP\nM=M-1\nA=M\nD=!M\nM=D\n@SP\nM=M+1\n')

    def other(self, line,i):
        
        pass


    #function to find words in a string
    def find_words(self, text, search):
        """Find exact words"""
        dText   = text.split()  
        dSearch = search.split()

        found_word = 0

        for text_word in dText:
            for search_word in dSearch:
                if search_word == text_word:
                    found_word += 1

        if found_word == len(dSearch):
            return True
        else:
            return False

    def main(self):
        i=0
        lines = self.inputdata.readlines()
        for line in lines:
            if (line[0]) != '/':
                #print(self, line)
                #it creates a command in the .asm file
                self.asmFile = self.asmFile + ("//INSTRUCTION: "+line)
                i=i+1
                #it classifies the asm intruction

                # Compare the inst, with the massive array of array
                for instruction in self.instructions:
                    if self.find_words(line,instruction[0]):
                        # This part calls the function with the input params
                        instruction[2](line,i)
                #for char in line:
                    #print(char)
        #add (END)\n @END\n 0;JMP\n to the .asm file
        self.asmFile = self.asmFile + ('(END)\n')
        self.asmFile = self.asmFile + ('@END\n')
        self.asmFile = self.asmFile + ('0;JMP\n')
        self.inputdata.close
        print(self.asmFile)
        file = open(self.asmFileLoc, "w")
        file.write(self.asmFile)
```
